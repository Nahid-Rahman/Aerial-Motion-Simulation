```python
import numpy as np                              #numeric python
import matplotlib.pyplot as plt                 #matlab style plotting
import scipy.integrate as sci                   #integration toolbox
```

```python
mass = 640.0/1000.0                             #rocket kg
```

```python
def Derivatives(state,t):
    global mass
    
    z = state[0]                                #state vector
    velz = state[1]
    zdot = velz                                 #Compute zdot - Kinematic Relationship
    
    gravity = -9.81*mass
    aero = 0.0
    thrust = 0.0
    
    Forces = gravity + aero + thrust  
    zddot = Forces/mass                         #Compute Acceleration   
    statedot = np.asarray([zdot,zddot])         #Compute the statedot
    
    return statedot
```

```python
###Initial Conditions
z0 = 0.0                                    
velz0 = 164.0 #m/s
stateinitial = np.asarray([z0,velz0])
```

```python
##Time window
tout = np.linspace(0,35,1000)
```

```python
###Numerical Integration Call
stateout = sci.odeint(Derivatives,stateinitial,tout)

###REname variables
zout = stateout[:,0]
velzout = stateout[:,1]
```

```python
#Plot ALTITUDE
plt.plot(tout,zout)
plt.xlabel('Time (sec)')
plt.ylabel('Altitude (m)')
plt.grid()
```

```python
#Plot VELOCITY
plt.figure()
plt.plot(tout,velzout)
plt.xlabel('Time (sec)')
plt.ylabel('Normal Speed (m/s)')
plt.grid()
```
